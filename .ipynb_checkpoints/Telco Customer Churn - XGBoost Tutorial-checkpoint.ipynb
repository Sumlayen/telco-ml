```python
import pandas as pd #load and manipulate data and for One-Hot Encoding
import numpy as np #Calculate the mean and standard deviation
import xgboost as xgb #XGBoost Stuff
from sklearn.model_selection import train_test_split #Split data into training and testing sets
from sklearn.metrics import balanced_accuracy_score, roc_auc_score, make_scorer #For scoring during
from sklearn.model_selection import GridSearchCV #Cross Validation
from sklearn.metrics import confusion_matrix #Create confusion matrix
from sklearn.metrics import plot_confusion_matrix #Draws a confusion matrix
```

```python
df = pd.read_csv('Telco_customer_churn.csv')
df.head()
```

```python
df.drop(['Churn Label', 'Churn Score', 'CLTV', 'Churn Reason'], axis=1, inplace=True) #Set axis=0 to remove rows, axis=1 to remove columns
df.head()
```

```python
df.drop(['CustomerID', 'Count', 'Country', 'State', 'Lat Long'], axis=1, inplace=True)
df.head()
```

```python
df['City'].replace(' ', '_', regex=True, inplace=True)
df.head()
```

```python
df.columns = df.columns.str.replace(' ', '_')
df.head()
```

```python
df.dtypes
```

```python
len(df.loc[df['Total_Charges'] == ' '])
```

```python
df.loc[df['Total_Charges'] == ' ']
```

```python
df.loc[(df['Total_Charges'] == ' '), 'Total_Charges'] = 0
```

```python
df.loc[df['Tenure_Months'] == 0]
```

```python
df['Total_Charges'] = pd.to_numeric(df['Total_Charges'])
df.dtypes
```

```python
df.replace(' ', '_', regex=True, inplace=True)
df.head()
```

```python
X = df.drop('Churn_Value', axis=1).copy() #Alternatively: X = df_no_missing.iloc[:,:-1]
X.head()
```

```python
y = df['Churn_Value'].copy()
y.head()
```

```python
X.dtypes
```

```python
pd.get_dummies(X, columns=['Payment_Method']).head()
```

```python
X_encoded = pd.get_dummies(X, columns = ['City',
                                        'Gender',
                                        'Senior_Citizen',
                                        'Partner',
                                        'Dependents',
                                        'Phone_Service',
                                        'Multiple_Lines',
                                        'Internet_Service',
                                        'Online_Security',
                                        'Online_Backup',
                                        'Device_Protection',
                                        'Tech_Support',
                                        'Streaming_TV',
                                        'Streaming_Movies',
                                        'Contract',
                                        'Paperless_Billing',
                                        'Payment_Method'])
X_encoded.head()
```

```python
sum(y)/len(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, random_state=42, stratify=y)
```

```python
sum(y_train)/len(y_train)
```

```python
sum(y_test)/len(y_test)
```

```python
clf_xgb = xgb.XGBClassifier(objective='binary:logistic', seed=42)
clf_xgb.fit(X_train,
           y_train,
           verbose=True,
           early_stopping_rounds=10,
           eval_metric='aucpr',
           eval_set=[(X_test, y_test)])
```

```python
#Broken***************
plot_confusion_matrix(clf_xgb,
                     X_test,
                     y_test,
                     values_format='d',
                     display_labels=["Did not leave", "Left"])
```
